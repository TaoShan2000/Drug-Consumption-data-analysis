```r
df0 <- read.csv("/content/dfnew.csv")
df0 <- data.frame(df0)
```

```r
# bar chart
ggplot(df0, aes(x = Cannabis)) +
  geom_bar(fill = "lightblue")
```

```r
f18 <- filter(df0, Age == -0.95197)
f25 <- filter(df0, Age == -0.07854)
f35 <- filter(df0, Age == 0.49788)
f45 <- filter(df0, Age == 1.09449)
f55 <- filter(df0, Age == 1.82213)
f65 <- filter(df0, Age == 2.59171)
```

```r
Age_18_count <- count(f18, Cannabis)
Age_18_percent <- Age_18_count[, 2] / sum(Age_18_count[, 2])

Age_25_count <- count(f25, Cannabis)
Age_25_percent <- Age_25_count[, 2] / sum(Age_25_count[, 2])

Age_35_count <- count(f35, Cannabis)
Age_35_percent <- Age_35_count[, 2] / sum(Age_35_count[, 2])

Age_45_count <- count(f45, Cannabis)
Age_45_percent <- Age_45_count[, 2] / sum(Age_45_count[, 2])

Age_55_count <- count(f55, Cannabis)
Age_55_percent <- Age_55_count[, 2] / sum(Age_55_count[, 2])

Age_65_count <- count(f65, Cannabis)
Age_65_percent <- rep(0, 7)
Age_65_percent[1] <- Age_65_count[1, 2] / sum(Age_65_count[, 2])
Age_65_percent[2] <- Age_65_count[2, 2] / sum(Age_65_count[, 2])
Age_65_percent[4] <- Age_65_count[3, 2] / sum(Age_65_count[, 2])
Age_65_percent
```

```r
percent <- c(Age_18_percent, Age_25_percent,Age_35_percent, Age_45_percent, Age_55_percent, Age_65_percent)
Age <- c(rep(0,7), rep(1,7), rep(2,7), rep(3,7), rep(4,7), rep(5,7))
Cannabis <- c(0:6, 0:6, 0:6, 0:6, 0:6, 0:6)
Age_Cannabis_plot <- cbind(Age, percent, Cannabis)
Age_Cannabis_plot <- data.frame(Age_Cannabis_plot)
```

```r
# Cannabis - age - count
Age_Cannabis_plot$Age <- as.factor(Age_Cannabis_plot$Age)
ggplot(Age_Cannabis_plot, aes(x = Cannabis, y = percent, fill = Age)) +
  geom_col(position = "dodge") +
  scale_fill_discrete(labels=c("18-24", "25-34", "35-44",
                               "45-54", "55-64", "65+"))
```
